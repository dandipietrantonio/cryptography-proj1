# Character Distribution Analysis

```python
def count_char(text):
    keys = [chr(ord('a')+i) for i in range(26)]
    keys.insert(0, ' ')
    distribution = dict(zip(keys,[0 for _ in range(27)]))
    for c in text:
        distribution[c] += 1
    return distribution
```

```python
keys = [chr(ord('a')+i) for i in range(26)]
keys.insert(0, ' ')
aggregrated_distribution = dict(zip(keys,[0 for _ in range(27)]))
```

```python
def show_distribution(text, aggre):
    distribution = count_char(text)
    if aggre:
        for key, val in distribution.items():
            aggregrated_distribution[key] += val
    plt.figure(figsize=(9,7))
    plt.bar(distribution.keys(), distribution.values())
    plt.show()
```

```python
for text in plaintexts:
    show_distribution(text, True)
```

# Vocabulary Analysis

```python
vocabulary_fre = {}
for text in plaintexts:
    for word in text.split(' '):
        if vocabulary_fre.get(word) is None:
            vocabulary_fre[word] = 1
        else:
            vocabulary_fre[word] += 1
```

```python
for key, val in vocabulary_fre.items():
    if val > 1:
        print(key, val)
```
